```python

n = np.array(range(0,20))
y1 = 0 * n + 1
y2 = np.log(n)
y3 = n
y4 = n * np.log(n)
y5 = n**2
y6 = 2**n
y7 = [math.factorial(i) for i in n]
y8 = n**n

plt.figure(figsize=(18,8))
y1_label = '$O(1)$'
y2_label = '$O(log n)$'
y3_label = '$O(n)$'
y4_label = '$O(n log n)$'
y5_label = '$O(n^2)$'
y6_label = '$O(2^n)$'
y7_label = '$O(n!)$'
y8_label = '$O(n^n)$'


plt.plot(n, y1,label = y1_label)
plt.plot(n, y2,label = y2_label)
plt.plot(n, y3,label = y3_label)
plt.plot(n, y4,label = y4_label)
plt.plot(n, y5,label = y5_label)
plt.plot(n, y6,label = y6_label)
plt.plot(n, y7,label = y7_label)
plt.plot(n, y8,label = y8_label)
plt.legend()

#plt.xlim(1,19)
#plt.ylim(0,50)

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.show()
```
